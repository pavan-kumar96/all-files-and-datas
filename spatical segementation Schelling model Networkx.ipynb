```python
#Base code for the schelling model
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 10
G = nx.grid_2d_graph(N,N)
```

```python
nx.draw(G)
```

```python
#Since the graph is not in a good format , we have get pos of the all the nodes in the data
G.nodes()
```

```python
#Since the nodes are in tuple of len two, first value is row and second value represents the column
pos = dict([(n,n) for n in G.nodes()])
```

```python
nx.draw(G, pos)
```

```python
# Since we now that labels are missing we want add the labels here is how
nx.draw_networkx(G, pos= pos, with_labels= True)
```

```python
#Since the above graph is weird.
labels = dict([((i,j), i*10+j) for i,j in  G.nodes()])
nx.draw_networkx(G,labels = labels, pos = pos)
```

```python
for (i,j) in G.nodes():
    print(i,j)
```

```python
for (i,j) in G.nodes():
    if (i+1 <= N-1) and (j+1  <= N-1):
        G.add_edge(i+1, j+1)
   
```
